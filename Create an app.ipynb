```python
import tkinter as tk
import random

main = tk.Tk()
main.title("Did you enjoy this semester")
main.geometry("400x250")
main.resizable(False, False)

user_feedback = {
    "likes_work": None,
    "excited_job": None,
    "rating": None
}

all_ratings = []

def show_summary():
    for widget in main.winfo_children():
        widget.destroy()

    total = len(all_ratings)
    average = sum(all_ratings) / total if total > 0 else 0

    summary_text = f"Thank you!\n\nTotal Responses: {total}\nAverage Rating: {average:.2f} ⭐"
    label_summary = tk.Label(main, text=summary_text, font=("Arial", 12), fg="black")
    label_summary.pack(pady=20)

    btn_quit = tk.Button(main, text="Exit", command=main.destroy)
    btn_quit.pack()

def submit_rating(rating):
    user_feedback["rating"] = rating
    all_ratings.append(rating)
    print("Survey Result:", user_feedback)
    show_summary()

def show_rating_screen():
    for widget in main.winfo_children():
        widget.destroy()

    label3 = tk.Label(main, text="Please rate your experience in this class (1 to 5 stars)", font=("Arial", 12), fg="black")
    label3.pack(pady=10)

    for i in range(1, 6):
        star_button = tk.Button(main, text="★" * i, font=("Arial", 14), command=lambda r=i: submit_rating(r))
        star_button.pack(pady=2)

def on_yes_2():
    user_feedback["excited_job"] = "Yes"
    show_rating_screen()

def on_no_2():
    user_feedback["excited_job"] = "No"
    show_rating_screen()

def on_yes():
    user_feedback["likes_work"] = "Yes"

    for widget in main.winfo_children():
        widget.destroy()

    label2 = tk.Label(main, text="Did you enjoy the assignments?", font=("BOLD", 14), fg="purple")
    label2.pack(pady=20)

    btn_yes_2 = tk.Button(main, text="Yes", fg="green", command=on_yes_2)
    btn_yes_2.pack(side="left", padx=40)

    btn_no_2 = tk.Button(main, text="No", fg="red", command=on_no_2)
    btn_no_2.pack(side="right", padx=40)

def on_no():
    user_feedback["likes_work"] = "No"

    main.update_idletasks()
    max_x = main.winfo_width() - button_no.winfo_reqwidth()
    max_y = main.winfo_height() - button_no.winfo_reqheight()
    new_x = random.randint(0, max(0, max_x))
    new_y = random.randint(40, max(40, max_y))
    button_no.place(x=new_x, y=new_y)

label = tk.Label(main, text="DID YOU HAVE FUN?", font=("BOLD", 16), fg="blue")
label.place(x=40, y=10)

button_yes = tk.Button(main, text="YES!!!", fg="green", cursor="circle", command=on_yes)
button_yes.place(x=80, y=60)

button_no = tk.Button(main, text="NO!!!", fg="red", cursor="pirate", command=on_no)
button_no.place(x=180, y=60)

main.mainloop()
```
